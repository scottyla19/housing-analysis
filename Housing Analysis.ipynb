```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# df = pd.read_csv('Zip_MedianValuePerSqft_AllHomes.csv')
df= pd.read_csv('http://files.zillowstatic.com/research/public/Zip/Zip_MedianValuePerSqft_AllHomes.csv')
df.tail()
```

```python
waco = df[df["Metro"] == "Waco"]
waco.head(10)
```

```python
dates = waco.columns[7:]
plot_table = pd.melt(waco.reset_index(), value_vars=waco.columns[7:], id_vars=["City"])
plot_table
fig, ax = plt.subplots(figsize=(15,10))
ax.set_xticks([])
ax.set_xlabel("Time")
ax.set_ylabel("Median Price per sq ft.")

g = sns.pointplot(data = plot_table, x = 'variable', y = "value", hue="City", markers=".")
```
