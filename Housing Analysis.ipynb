```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
sns.set()

%matplotlib inline
```

```python
# df = pd.read_csv('Zip_MedianValuePerSqft_AllHomes.csv')
df= pd.read_csv('http://files.zillowstatic.com/research/public/Zip/Zip_MedianValuePerSqft_AllHomes.csv')
df.tail()
```

```python
waco = df[df["Metro"] == "Waco"]

waco.head(10)
```

```python

plot_table = pd.melt(waco.reset_index(), value_vars=waco.columns[7:], id_vars=["City"])
dfYrM = pd.DataFrame(plot_table['variable'].str.split('-').tolist(), columns=['Year','Month'])
plot_table["Year"] = dfYrM["Year"]
plot_table["Month"] = dfYrM["Month"]
plot_table

```

## Time Series

```python
fig, ax = plt.subplots(figsize=(15,10))

g = sns.pointplot(data = plot_table, x = 'variable', y = "value", hue="City", markers=".", scale = .5)
g.set_xticks([])
g.set_xlabel("Time")
g.set_ylabel("Median Price per sq ft.")
```

```python
g= sns.FacetGrid(plot_table, col='Month', row="City")
g = g.map(plt.plot, "Year", "value")
g.set(xticks=plot_table.Year[2::8])
# g.set_xticklabels([])


# for ax in g.axes.flat:
#     for label in ax.xaxis.get_ticklabels():
#         label.set_visible(False)
```
